```python
# Your FID metric values
epochs = [50, 100, 150, 200]
fid_values_central = [279.0527474097098, 208.85317152042427, 209.5353934053711, 190.9847592153065]
fid_values_site = [340.7764292350358, 247.66225086937212, 265.45961268671954, 259.1145972171544]
fid_values_federated = [282.8523129396619, 220.21262486503966, 236.63588280160866, 206.08029278701636]
fid_values_dpfed = [274.5464297143294, 242.13335586010686, 227.41657590365855, 215.11081269729948]

plt.plot(epochs, fid_values_central, linestyle='-', marker='o', markersize=8, label='Central')
plt.plot(epochs, fid_values_site, linestyle='-', marker='o', markersize=8, label='Site')
plt.plot(epochs, fid_values_federated, linestyle='-', marker='o', markersize=8, label='Federated')
plt.plot(epochs, fid_values_dpfed, linestyle='-', marker='o', markersize=8, label='Federated DP')

plt.xlabel('Epochs')
plt.ylabel('FID Metric')
plt.title('FID Metric vs. Epochs')

# Customize the plot's appearance (optional)
# plt.grid(True)
plt.xticks(epochs)
plt.legend()

# Save the plot as a file (optional)
plt.savefig('fid.png')

plt.show()
```
