```python
import os

files = os.listdir("/Users/eduardpauliuc/PycharmProjects/federated/maps_all/val")
# print(files)

mapped = {}

for idx, name in enumerate(files):
    oldn = name[:-4]
    newn = idx

    mapped[str(newn)] = oldn

print(mapped)
```

```python
gen_dir = '/Users/eduardpauliuc/PycharmProjects/federated/generated/site_gen_200'
dest_dir = '/Users/eduardpauliuc/PycharmProjects/federated/generated/site_gen_200_good'

import shutil

if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

for idx, val in enumerate(sorted(os.listdir(gen_dir), key = lambda f: int(f[:-4]))):
    newn = mapped[val[:-4]] + '.png'
    print(idx, val, newn)
    old_path = os.path.join(gen_dir, val)
    new_path = os.path.join(dest_dir, newn)
    shutil.copy2(old_path, new_path)
```
